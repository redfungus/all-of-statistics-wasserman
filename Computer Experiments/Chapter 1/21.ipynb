```python
import numpy as np
import altair as alt
import pandas as pd
```

```python
def coin_toss(n, p):
    return [np.random.binomial(1,p) for _ in range(n)]
```

```python
def get_coin_toss_data(n, p):
    heads_tails = coin_toss(n, p)
    heads_proportion = [sum(heads_tails[:i])/(i+1) for i in range(n)]

    data = pd.DataFrame({
        'x': [i+1 for i in range(n)],
        'y': heads_proportion
        })

    return data


```

```python
# Number of coin tosses.
n = 1000
# Probability of heads
p = 0.3

alt.Chart(get_coin_toss_data(n, p)).mark_line().encode(
        x='x',
        y='y',
    )
```

```python
# Number of coin tosses.
n = 1000
# Probability of heads
p = 0.03

alt.Chart(get_coin_toss_data(n, p)).mark_line().encode(
        x='x',
        y='y',
    )
```
