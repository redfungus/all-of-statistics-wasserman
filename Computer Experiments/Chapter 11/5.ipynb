```python
import numpy as np
import altair as alt
import pandas as pd
import scipy
from scipy.stats import beta
import math
alt.data_transformers.disable_max_rows()
```

```python
x = np.linspace(0, 1, 200)
```

```python
n = 10
s = 2
```

```python
paramaters = [
    (0.5, 0.5),
    (1, 1),
    (10, 10),
    (100, 100),
]
```

```python
colors = ['red', 'green', 'blue', 'black']
```

```python
charts = []
for distribution_params, color in zip(paramaters, colors):
    posterior_density_data = pd.DataFrame({
        'x': x,
        'y': beta.pdf(x, s + distribution_params[0], n - s + distribution_params[1])
        })

    posterior_density_chart = alt.Chart(posterior_density_data).mark_line(filled=False, color=color).encode(
    alt.X('x'),
    alt.Y('y')
    )

    charts.append(posterior_density_chart)
```
