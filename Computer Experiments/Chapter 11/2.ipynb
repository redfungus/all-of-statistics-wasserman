```python
import numpy as np
import altair as alt
import pandas as pd
import scipy
from scipy.stats import norm, lognorm
alt.data_transformers.disable_max_rows()
```

```python
mu = 5
```

```python
sigma = 1
```

```python
n = 100
```

```python
data = np.random.normal(mu, sigma, size=100)
```

```python
# Posterior ~ N(X_bar, 1/n)
x = np.linspace(4.98, 5.08, 100)
```

```python
posterior_density_data = pd.DataFrame({
        'x': x,
        'y': norm.pdf(x, np.mean(data), (sigma**2)/n)
        })
```

```python
posterior_density_chart = alt.Chart(posterior_density_data).mark_line(filled=False, color="red").encode(
    alt.X('x'),
    alt.Y('y')
)
```

```python
# Posterior ~ N(X_bar, 1/n)
posterior_data_simulated = np.random.normal(np.mean(data), (sigma**2)/n, size=1000)
```

```python
histogram_data = pd.DataFrame({
        'x': posterior_data_simulated,
        })
```

```python
posterior_histogram_chart = alt.Chart(histogram_data).transform_density(
    'x',
    as_=['x', 'density']
).mark_line(filled=False, color="blue").encode(
    alt.X('x:Q'),
    alt.Y('density:Q')
)
```
