```python
import numpy as np
import altair as alt
import pandas as pd
from scipy.stats import norm
import math
```

a)

```python
X = np.arange(0.01, 10, step=0.01)
Y = [norm.pdf(np.log(x))/x for x in X]
```

```python
data = pd.DataFrame({
        'x': X,
        'y': Y
        })
```

```python
true_chart = alt.Chart(data).mark_line(color='red').encode(
        x=alt.X('x', axis=alt.Axis( title='y')),
        y=alt.Y('y', axis=alt.Axis(title='f(y)')),
    )

true_chart
```

b)

```python
X = []
Y = []

for i in range(0, 10_000):
    X.append(np.random.normal())
    Y.append(math.exp(X[i]))
```

```python
data = pd.DataFrame({
        'x': Y
        })
```

```python
alt.data_transformers.disable_max_rows()

experiment_chart = alt.Chart(data).transform_joinaggregate(
    total='count(*)'
).transform_calculate(
    pct='1 / datum.total'
).mark_bar(filled=False).encode(
    alt.X('x', bin=alt.Bin(step=1)),
    alt.Y('sum(pct):Q', axis=alt.Axis(format='%'))
)

experiment_chart
```

```python
true_chart + experiment_chart
```
