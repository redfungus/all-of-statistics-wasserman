```python
import numpy as np
import altair as alt
import random
alt.data_transformers.disable_max_rows()
```

```python
a = 0.1
b = 0.3
```

```python
P = np.array([
    [1-a, a], 
    [b, 1-b]
])
```

```python
initial_state = 0
```

```python
simulations = 10000
```

```python
states = [initial_state]
for _ in range(1, simulations):
    previous_state = states[-1]
    new_state = 0 if random.random() < P[previous_state][0] else 1
    states.append(new_state)

states = np.array(states)
```

```python
p1 = np.cumsum(states == 0) / np.arange(1, simulations + 1)
p2 = np.cumsum(states == 1) / np.arange(1, simulations + 1)
```
