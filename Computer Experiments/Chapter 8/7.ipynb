```python
import numpy as np
import altair as alt
import pandas as pd
import scipy
from scipy.stats import norm, lognorm
import math
alt.data_transformers.disable_max_rows()
```

```python
n = 50
```

```python
real_theta = 1
```

```python
B = 1000
```

```python
data = np.random.uniform(0, real_theta, n)
```

```python
theta_hat = max(data)
```

```python
bootstraps = []
for i in range(B):
    theta_hat_b = max(np.random.choice(data, replace=True, size=len(data)))
    bootstraps.append(theta_hat_b)
```

```python
se = math.sqrt(np.var(bootstraps))
```

```python
print("Estimated Standard Error:", se)
```

```python
histogram_data = pd.DataFrame({
        'x': bootstraps,
        })
```

```python
histogram_chart = alt.Chart(histogram_data).transform_density(
    'x',
    as_=['x', 'density']
).mark_line(filled=False).encode(
    alt.X('x:Q'),
    alt.Y('density:Q')
)
```

```python
histogram_chart
```

```python
real_distribution = lambda x: n*(x**(n-1))
```

```python
data_range = np.linspace(0.8, 1, 100)
```

```python
true_histogram_data = pd.DataFrame({
        'x': data_range,
        'y': list(map(real_distribution, data_range))
        })
```

```python
true_histogram_chart = alt.Chart(true_histogram_data).mark_line(filled=False).encode(
    alt.X('x'),
    alt.Y('y')
)
```
