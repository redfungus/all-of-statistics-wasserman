```python
import numpy as np
import altair as alt
import pandas as pd
from scipy.stats import norm
import math
```

```python
n_max = 10000
```

```python
X_bars_normal = [sum(np.random.normal(size=n))/n for n in range(1, n_max+1)]
```

```python
data = pd.DataFrame({
        'x': range(1, n_max+1),
        'y': X_bars_normal
        })
```

```python
alt.data_transformers.disable_max_rows()

chart_normal = alt.Chart(data).mark_line().encode(
        x=alt.X('x', axis=alt.Axis(title='Number of samples')),
        y=alt.Y('y', axis=alt.Axis(title='Sample Average')),
    )
```

```python
chart_normal
```

```python
X_bars_cauchy = [sum(np.random.standard_cauchy(size=n))/n for n in range(1, n_max+1)]
```

```python
data = pd.DataFrame({
        'x': range(1, n_max+1),
        'y': X_bars_cauchy
        })
```

```python
alt.data_transformers.disable_max_rows()

chart_cauchy = alt.Chart(data).mark_line(color="red").encode(
        x=alt.X('x', axis=alt.Axis(title='Number of samples')),
        y=alt.Y('y', axis=alt.Axis(title='Sample Average')),
    )
```

```python
chart_cauchy
```

```python
chart_normal | chart_cauchy
```
