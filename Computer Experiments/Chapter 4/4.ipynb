```python
import numpy as np
import altair as alt
import pandas as pd
from scipy.stats import uniform
import math
alt.data_transformers.disable_max_rows()
```

```python
alpha = 0.05
```

```python
p = 0.4
```

```python
n_values = range(1, 10_000+1)
```

```python
simulations = 50000
```

```python
coverages = []
for n in n_values:
    epsilon = math.sqrt(math.log(2/alpha)/(2*n))
    p_n_hats = np.random.binomial(n, p, size=simulations) / n
    coverages.append(
        sum([p_n_hat - epsilon <= p_n_hat <= p_n_hat + epsilon for p_n_hat in p_n_hats]) / simulations
    )
```

```python
data = pd.DataFrame({
        'x': n_values,
        'y': coverages
        })
```

```python
chart = alt.Chart(data).mark_line().encode(
        x=alt.X('x', axis=alt.Axis( title='y')),
        y=alt.Y('y', axis=alt.Axis(title='f(y)'),  scale=alt.Scale(
            domain=(0.9, 1.1),
            clamp=True
        )),
    )

chart
```

```python
interval_length_data = pd.DataFrame({
        'x': n_values,
        'y': [2 * math.sqrt(math.log(2/alpha)/(2*n)) for n in n_values]
        })
```

```python
interval_length_chart = alt.Chart(interval_length_data).mark_line().encode(
        x=alt.X('x', axis=alt.Axis( title='y')),
        y=alt.Y('y', axis=alt.Axis(title='f(y)')),
    )

interval_length_chart
```
