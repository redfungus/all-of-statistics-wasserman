```python
import numpy as np
import altair as alt
import pandas as pd
import scipy
from scipy.stats import norm, lognorm
import math
alt.data_transformers.disable_max_rows()
```

```python
n = 50
```

```python
real_theta = 1
```

```python
B = 100
```

```python
data = np.random.uniform(0, real_theta, n)
```

```python
theta_hat = max(data)
```

```python
nonparametric_bootstraps = []
for i in range(B):
    theta_hat_b = max(np.random.choice(data, replace=True, size=len(data)))
    nonparametric_bootstraps.append(theta_hat_b)
```

```python
non_parametric_se = math.sqrt(np.var(nonparametric_bootstraps))
```

```python
print("Estimated Standard Error non-parametric:", non_parametric_se)
```

```python
parametric_bootstraps = []
for i in range(B):
    theta_hat_b = max(np.random.uniform(0, theta_hat, size=len(data)))
    parametric_bootstraps.append(theta_hat_b)
```

```python
parametric_se = math.sqrt(np.var(parametric_bootstraps))
```

```python
print("Estimated Standard Error parametric:", parametric_se)
```

```python
nonparametric_histogram_data = pd.DataFrame({
        'x': nonparametric_bootstraps,
        })

parametric_histogram_data = pd.DataFrame({
        'x': parametric_bootstraps,
        })
```

```python
non_parametric_histogram_chart = alt.Chart(nonparametric_histogram_data).transform_density(
    'x',
    as_=['x', 'density']
).mark_line(filled=False, color="blue").encode(
    alt.X('x:Q'),
    alt.Y('density:Q')
)

parametric_histogram_chart = alt.Chart(parametric_histogram_data).transform_density(
    'x',
    as_=['x', 'density']
).mark_line(filled=False, color="red").encode(
    alt.X('x:Q'),
    alt.Y('density:Q')
)
```

```python
non_parametric_histogram_chart + parametric_histogram_chart
```

```python
real_distribution = lambda x: n*(x**(n-1))
```

```python
data_range = np.linspace(0.8, 1, 100)
```

```python
true_histogram_data = pd.DataFrame({
        'x': data_range,
        'y': list(map(real_distribution, data_range))
        })
```

```python
true_histogram_chart = alt.Chart(true_histogram_data).mark_line(filled=False, color="black").encode(
    alt.X('x'),
    alt.Y('y')
)
```

```python
true_histogram_chart + non_parametric_histogram_chart + parametric_histogram_chart
```
